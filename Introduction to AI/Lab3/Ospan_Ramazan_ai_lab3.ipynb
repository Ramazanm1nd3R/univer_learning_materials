```python
#10.String Operations:
#Extract the last name of each passenger into a new column called LastName.
titanic_df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
titanic_df['LastName'] = titanic_df['Name'].apply(lambda x : x.split(',')[0])
titanic_df
```

```python
#Create a new column Title by extracting titles (like Mr, Mrs, Miss) from the Name column.
def titleCheck(row):
    titles = ['Mr', 'Mrs', 'Miss']
    words = row['Name'].split()
    for word in words:
        if word.strip('.') in titles:
            return word.strip('.')

    if row['Sex'] == 'male':
        return 'Mr'
    elif row['Sex'] == 'female':
        return 'Miss'
    return None

titanic_df['Title'] = titanic_df.apply(titleCheck, axis = 1)
titanic_df
```

```python
#11. Conditional Columns:
#Create a new column AgeGroup that categorizes passengers as 'Child', 'Adult', or 'Senior' based on their age
def ageCheck (age):
    if age >= 50:
        return 'Senior'
    if 18 <= age <= 49:
        return 'Adult'
    return 'Child'

titanic_df = titanic_df.loc[titanic_df['Age'].notna()]
titanic_df['AgeGroup'] = titanic_df['Age'].apply(ageCheck)
titanic_df
```

```python
#Create a new column IsAlone which is True if FamilySize is 0, otherwise False.

titanic_df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')

# (from other task) Create a new column called FamilySize which is the sum of SibSp and Parch.
titanic_df['FamilySize'] = titanic_df['SibSp'] + titanic_df['Parch']


titanic_df['IsAlone'] = titanic_df['FamilySize'].apply(lambda x: x == 0)
titanic_df
```

```python
#12.Merging DataFrames:
#Create a separate dataframe with the columns PassengerId and Survived. Merge this dataframe back with the original dataframe on PassengerId.
titanic_df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')

new_df = titanic_df[['PassengerId', 'Survived']]

merged_df = titanic_df.merge(new_df, how='inner', on='PassengerId')
merged_df = merged_df.drop(columns=['Survived_y'])
merged_df.rename(columns={'Survived_x': 'Survived'}, inplace=True)
merged_df
```

```python
#13.Pivot Table:
#Create a pivot table showing the average Fare paid for each combination of Pclass and Embarked.
titanic_df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
groupped_df = titanic_df.groupby(['Pclass', 'Embarked'])['Fare'].mean().round(2)
groupped_df
```

```python
#14.Crosstab:
#Create a crosstab showing the count of passengers who survived versus those who did not for each Sex
crosstab_df = pd.crosstab(titanic_df['Sex'], titanic_df['Survived'])
crosstab_df
```

```python
"""Iris Dataset Overview"""
# (0 = Setosa, 1 = Versicolor, 2 = Virginica).

# датасет не был предоставлен, нашел сам в интернете

iris_df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data')
iris_df.columns = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)', 'target']
iris_df
```

```python
#2. Data Selection:
#Select the columns sepal length (cm) and sepal width (cm) and display the first 5 rows.

dfdf_df = iris_df[['sepal length (cm)', 'sepal width (cm)']].head(5)
dfdf_df
```

```python
#4. Aggregation and Grouping:
#Calculate the average petal length (cm) for each species.
groupped_df = iris_df.groupby('target')['petal length (cm)'].mean().round(2)
groupped_df
```

```python
#5. Creating New Columns:
#Create a new column called sepal_area which is the product of sepal length (cm) and sepal width (cm).
iris_df['sepal_area'] = iris_df['sepal length (cm)'] * iris_df['sepal width (cm)']
iris_df
```

```python
#Create a new column petal_area which is the product of petal length (cm) and petal width (cm).
iris_df['petal_area'] = iris_df['petal length (cm)'] * iris_df['petal width (cm)']
iris_df
```

```python
#6. Handling Categorical Data:
#Replace the numerical values in the target column with the actual species names: 0 for 'Setosa', 1 for 'Versicolor', and 2 for 'Virginica'.

#ну, в датасете который я нашел итак названия, так что сделаю наоборот

def changeNames (aaa):
    if aaa == 'Iris-setosa':
        return 0
    if aaa == 'Iris-versicolor':
        return 1
    if aaa == 'Iris-virginica':
        return 2

iris_df['target'] = iris_df['target'].apply(changeNames)
iris_df
```

```python
#Create a new column species_type that categorizes the species as 'Small', 'Medium', or 'Large' based on the average petal length (cm).

avgPetalLength = iris_df.groupby('species')['petal length (cm)'].mean()

def checkType(length):
    if length > 1.8:
        return 'Large'
    elif length >= 1.2:
        return 'Medium'
    else:
        return 'Small'

iris_df['species_type'] = iris_df['species'].apply(lambda species: checkType(avgPetalLength[species]))
iris_df
```

```python
#7. Sorting Data:
#Sort the dataset by sepal length (cm) in ascending order.
iris_df.sort_values(by = 'sepal length (cm)')
```

```python
#Sort the dataset by target and then by petal width (cm) in descending order
iris_df.sort_values(by = ['target', 'petal width (cm)'], ascending= [True, False])
```

```python
#8. Value Counts and Unique Values:
#Find the number of unique values in the target column.
uniqueCount = iris_df['target'].nunique()
uniqueCount
```

```python
#Get the count of each unique value in the sepal length (cm) column.
countOfAllSepalLengthes = iris_df.groupby('sepal length (cm)')['target'].count()
countOfAllSepalLengthes
```

```python
#10.String Operations:
#Create a new column species_code which takes the first three letters of the species name.
def namedNumbersOfIrises (target):
    if target == 0:
        return 'Setosa'
    if target == 1:
        return 'Versicolor'
    if target == 2:
        return 'Virginica'
#тут я заново создаю таблицу с названиями потому что ранее заменил все названия на цифры
iris_df['species'] = iris_df['target'].apply(namedNumbersOfIrises)

#написал через ляюбду потому что простая логика которую нужно было применить на все элементы
iris_df['species_code'] = iris_df['species'].apply(lambda x: x[:3])
iris_df
```

```python
#11. Conditional Columns:
#Create a new column flower_size that categorizes flowers as 'Small', 'Medium', or 'Large' based on the petal length (cm).

def checkType(length):
    if length > 1.8:
        return 'Large'
    elif length >= 1.2:
        return 'Medium'
    else:
        return 'Small'

iris_df['flower_size'] = iris_df['petal length (cm)'].apply(checkType)
iris_df
```

```python
#12.Statistical Analysis:
#Calculate the correlation matrix for the numerical columns.

#для того что бы вытащить только числовые таблицы
numColumns = iris_df.select_dtypes(include=['float64', 'int64'])

corrMatrix = numColumns.corr()
corrMatrix
```

```python
#14.Crosstab:
#Create a crosstab showing the count of each target for different sepal length (cm) ranges.

def lables(length):
    if length < 5.0:
        return '4-5'
    elif length < 6.0:
        return '5-6'
    elif length < 7.0:
        return '6-7'
    else:
        return '7-8'
iris_df['label'] = iris_df['sepal length (cm)'].apply(lables)
groupped_aaooaoaoa = iris_df.groupby(['label', 'target'])['sepal length (cm)'].count().unstack() #unstack() для более удобного просмотра
groupped_aaooaoaoa
```
